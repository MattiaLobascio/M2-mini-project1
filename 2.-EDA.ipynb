Finally, we will calculate the linear regression between the price and the weight. This will be done first by plotting it with the `seaborn` library and then calculating the error with the `scipy` library.

```python
# your code here
sns.regplot("carat","price",data=diamonds, truncate=True).set(ylim=(0,22000))
```

```python
from scipy import stats as st

X = diamonds['carat']
Y = diamonds['price']
slope, intercept, r_value, _, std_err = st.linregress(X, Y)
st.linregress(X, Y)
```

```python
#Find predicted values based on regression
Yp = slope*X + intercept
#Residuals as difference between predicted and real values
e = (Y - Yp)
RMSE = round(((sum(e**2)/len(e))**0.5),2)
print(RMSE)
print(diamonds["price"].min(),diamonds["price"].max())
```
